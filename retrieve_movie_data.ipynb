### Access the New York Times API

```python
# Set the base URL


url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = (
    f"{url}api-key={nyt_api_key}&begin_date={begin_date}&end_date={end_date}"
    + f'&fq={filter_query}&sort={sort}&fl={field_list}'
)

# response = requests.get(query_url,verify=False)
# response
# for i in range(4):
#   print(i)
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
reviews_list_norm_df = pd.json_normalize(reviews_list)
reviews_list_norm_df
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

# reviews_list_norm_df["title"] = reviews_list_norm_df["headline.main"].apply(lambda st: st[st.find("'")+1:st.find("Review")])
reviews_list_norm_df["title"] = reviews_list_norm_df["headline.main"].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])

reviews_list_norm_df
```

```python
# Extract 'name' and 'value' from items in "keywords" column
import ast


def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # if isinstance(item, dict):  # Check if the item is a dictionary
            # Extract 'name' and 'value'
            keyword = f"{item['name']}: {item['value']};"
            # Append the keyword item to the extracted_keywords string
            extracted_keywords += keyword
    return extracted_keywords


# Fix the "keywords" column by converting cells from a list to a string


reviews_list_norm_df['keywords']=reviews_list_norm_df['keywords'].apply(extract_keywords)
reviews_list_norm_df
# extracted_keywords = extract_keywords(reviews_list_norm_df_fixed['keywords'])
# reviews_list_norm_df_fixed
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database

reviews_list_norm_df["title"] = reviews_list_norm_df["title"].str.strip("'")
titles_list= reviews_list_norm_df["title"].to_list()
while("" in titles_list):
    titles_list.remove("")
titles_list
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merge_ny_tmd=reviews_list_norm_df.merge(results_df, on='title',how="inner")
merge_ny_tmd=merge_ny_tmd.set_index('title')
merge_ny_tmd=merge_ny_tmd.dropna(axis = 0, subset = ["headline.main","runtime","video","status"])
merge_ny_tmd
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
column_remove_list=["keywords","byline.person"]

# Create a list of characters to remove
char_remove_list=['[',']','\'']

merge_ny_tmd["keywords"]= merge_ny_tmd["keywords"].apply(str)
merge_ny_tmd["byline.person"]= merge_ny_tmd["byline.person"].apply(str)
# Loop through the list of columns to fix
for columnfix in column_remove_list:
    # Convert the column to type 'str'

    # # Loop through characters to remove
    for chrrmvlist in char_remove_list:

        merge_ny_tmd[columnfix] = merge_ny_tmd[columnfix].str.replace(chrrmvlist,'')
# Display the fixed DataFrame
merge_ny_tmd
```

```python
# Drop "byline.person" column
merge_ny_tmd=merge_ny_tmd.drop('byline.person',axis=1)
merge_ny_tmd

```

```python
# Delete duplicate rows and reset index

merge_ny_tmd = merge_ny_tmd.apply(lambda x: x.astype(str).str.lower()).drop_duplicates(keep=False) 
# merge_ny_tmd=merge_ny_tmd.drop_duplicates()
# merge_ny_tmd=merge_ny_tmd.reset_index(drop=False)
# merge_ny_tmd=merge_ny_tmd.set_index(['title'])

merge_ny_tmd
```
